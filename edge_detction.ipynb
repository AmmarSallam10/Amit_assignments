#sobel

```python
img=cv2.imread('edge.jpg')
```

```python
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
img=cv2.GaussianBlur(img,(3,3),0)
```

```python
sobely=cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)#y
sobelx=cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5)#x
```

```python
plt.imshow(img,cmap='gray')
plt.title('original')
plt.show()

plt.imshow(sobelx,cmap='gray')
plt.title('sobelx')
plt.show()

plt.imshow(sobely,cmap='gray')
plt.title('sobely')
plt.show()

plt.imshow(sobelx+ sobely,cmap='gray')
plt.title('sobelx+ sobely')
plt.show()
```

#prewitt

```python
img=cv2.imread('edge.jpg')
```

```python
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
img=cv2.GaussianBlur(img,(3,3),0)
```

```python
kernelx=np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
kernely=np.array([[-1,0,1],[-1,0,1],[-1,0,1]])

img_prewitt_x=cv2.filter2D(img,-1,kernelx)
img_prewitt_y=cv2.filter2D(img,-1,kernely)
```

```python
imshow(img)
```

```python
plt.imshow(img_prewitt_x,cmap='gray')
plt.title('img_prewitt_x')
plt.show()

plt.imshow(img_prewitt_y,cmap='gray')
plt.title('img_prewitt_y')
plt.show()

plt.imshow(img_prewitt_x+ img_prewitt_y,cmap='gray')
plt.title('img_prewitt_x+ img_prewitt_y')
plt.show()
```
